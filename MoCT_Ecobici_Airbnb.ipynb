## Ecobici Dataset

```python
file_path = '/content/drive/MyDrive/Colab 6 Semestre/Jesus/2025-01.csv'

# Leer CSV en DataFrame
df_ecobike = pd.read_csv(file_path)

#Mostrar una preview de los datos
print("Preview del dataframe")
display(df_ecobike.head(20))
```

```python
# Checamos los tipos de datos que tenemos
df_ecobike.info()
```

### Genero_Usuario

```python
# Moda
moda_genero = df_ecobike["Genero_Usuario"].mode()[0]
print("Género más frecuente:", moda_genero)
```

```python
#Número de usuarios por género
print("\nConteo de usuarios por género:")
print(df_ecobike["Genero_Usuario"].value_counts())

plt.figure(figsize=(6, 4))
sns.countplot(data=df_ecobike, x="Genero_Usuario", palette="viridis", hue="Genero_Usuario", legend=False)
plt.title("Distribución de Género")
plt.xlabel("Género")
plt.ylabel("Cantidad de Usuarios")
plt.show()
```

### Edad_Usuario

```python
# Media
media_edad = df_ecobike["Edad_Usuario"].mean()
print("Media Edad:", media_edad)

# Mediana
mediana_edad = df_ecobike["Edad_Usuario"].median()
print("Mediana Edad:", mediana_edad)

# Moda
moda_edad = df_ecobike["Edad_Usuario"].mode()[0]
print("Moda Edad:", moda_edad)
```

```python
# Histograma de las edades

plt.figure(figsize=(10,4))
sns.histplot(df_ecobike["Edad_Usuario"], alpha=0.4, bins=30, color="orange")

# Resaltar media, mediana y moda
plt.axvline(media_edad, color="red", linestyle="dashed", linewidth=2, label="Media")
plt.axvline(mediana_edad, color="green", linestyle="dashed", linewidth=2, label="Mediana")
plt.axvline(moda_edad, color="blue", linestyle="dashed", linewidth=2, label="Moda")

plt.text(media_edad, plt.gca().get_ylim()[1]*0.85, f'{media_edad:.1f}', color='red', rotation=90, ha='left')
plt.text(mediana_edad, plt.gca().get_ylim()[1]*0.75, f'{mediana_edad:.1f}', color='green', rotation=90, ha='left')
plt.text(moda_edad, plt.gca().get_ylim()[1]*0.65, f'{moda_edad}', color='blue', rotation=90, ha='right')


plt.title("Distribución de Edad de Usuarios")
plt.xlabel("Edad")
plt.ylabel("Cantidad de Usuarios")
plt.legend()
plt.show()
```

### Bici, Ciclo_Estacion_Retiro y Ciclo_EstacionArribo

```python
# Solo obtendremos la moda, pues la media y mediana no resultan útiles en este caso

# Bici
moda_bici = df_ecobike["Bici"].mode()[0]
print("Bicicleta más usada:", moda_bici)

# Convertimos las columnas de estaciones a tipo entero
df_ecobike["Ciclo_Estacion_Retiro"] = df_ecobike["Ciclo_Estacion_Retiro"].str.extract(r'(\d+)').astype(int)
df_ecobike["Ciclo_EstacionArribo"] = df_ecobike["Ciclo_EstacionArribo"].str.extract(r'(\d+)').astype(int)

# Ciclo Estacion Retiro
moda_estacion_retiro = df_ecobike["Ciclo_Estacion_Retiro"].mode()[0]
print("Estación con más retiros:", moda_estacion_retiro)

# Ciclo Estacion Arribo
moda_estacion_arribo = df_ecobike["Ciclo_EstacionArribo"].mode()[0]
print("Estación con más arribos:", moda_estacion_arribo)
```

### Fecha_Retiro y Fecha_Arribo


```python
# Primero convertimos a datetime
df_ecobike["Fecha_Retiro"] = pd.to_datetime(df_ecobike["Fecha_Retiro"], format="%d/%m/%Y")
df_ecobike["Fecha_Arribo"] = pd.to_datetime(df_ecobike["Fecha_Arribo"], format="%d/%m/%Y")

# Media
media_retiro_fecha = df_ecobike["Fecha_Retiro"].mean().date()
media_arribo_fecha = df_ecobike["Fecha_Arribo"].mean().date()
print("Media Fecha_Retiro:", media_retiro_fecha)
print("Media Fecha_Arribo:", media_arribo_fecha)

# Mediana
mediana_retiro_fecha = df_ecobike["Fecha_Retiro"].median().date()
mediana_arribo_fecha = df_ecobike["Fecha_Arribo"].median().date()
print("Mediana Fecha_Retiro:", mediana_retiro_fecha)
print("Mediana Fecha_Arribo:", mediana_arribo_fecha)

# Moda
moda_retiro_fecha = df_ecobike["Fecha_Retiro"].mode()[0].date()
moda_arribo_fecha = df_ecobike["Fecha_Arribo"].mode()[0].date()
print("Moda Fecha_Retiro:", moda_retiro_fecha)
print("Moda Fecha_Arribo:", moda_arribo_fecha)
```

```python
# ----- Histograma de fechas de retiro -----------
plt.figure(figsize=(10,5))
plt.hist(df_ecobike["Fecha_Retiro"], bins=pd.date_range("2025-01-01", "2025-02-01"), alpha=0.4, label="Fecha Retiro", edgecolor="black", color="skyblue")

# Resaltar media, mediana y moda
plt.axvline(media_retiro_fecha, color="red", linestyle="dashed", linewidth=3 , label="Media")
plt.axvline(mediana_retiro_fecha, color="green", linestyle="dashed", linewidth=2, label="Mediana")
plt.axvline(moda_retiro_fecha, color="blue", linestyle="dashed", linewidth=2, label="Moda")

plt.text(media_retiro_fecha, plt.gca().get_ylim()[1]*0.9, media_retiro_fecha.strftime('%d/%m/%Y'), color='red')
plt.text(mediana_retiro_fecha, plt.gca().get_ylim()[1]*0.85, mediana_retiro_fecha.strftime('%d/%m/%Y'), color='green', ha='right')
plt.text(moda_retiro_fecha, plt.gca().get_ylim()[1]*0.8, moda_retiro_fecha.strftime('%d/%m/%Y'), color='blue')

plt.title("Distribución de Fechas de Retiro - Enero 2025")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de viajes")
plt.legend()

# Formatear eje x como fechas
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%Y'))
plt.gcf().autofmt_xdate()

plt.show()
```

```python
# ----- Histograma de fechas de arribo -----------

plt.figure(figsize=(10,5))

plt.hist(df_ecobike["Fecha_Arribo"], bins=pd.date_range("2025-01-01", "2025-02-01"), alpha=0.4, edgecolor="black", color="orange")

plt.axvline(media_arribo_fecha, color="red", linestyle="dashed", linewidth=2, label="Media")
plt.axvline(mediana_arribo_fecha, color="green", linestyle="dashed", linewidth=2, label="Mediana")
plt.axvline(moda_arribo_fecha, color="blue", linestyle="dashed", linewidth=2, label="Moda")

plt.text(media_arribo_fecha, plt.gca().get_ylim()[1]*0.9, media_arribo_fecha.strftime('%d/%m/%Y'), color='red')
plt.text(mediana_arribo_fecha, plt.gca().get_ylim()[1]*0.85, mediana_arribo_fecha.strftime('%d/%m/%Y'), color='green', ha='right')
plt.text(moda_arribo_fecha, plt.gca().get_ylim()[1]*0.8, moda_arribo_fecha.strftime('%d/%m/%Y'), color='blue')

plt.title("Distribución de Fechas de Arribo - Enero 2025")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de viajes")
plt.legend()

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%Y'))
plt.gcf().autofmt_xdate()
plt.show()
```

### Hora_Retiro y Hora_Arribo

```python
# Primero convertimos a un formato adecuado (datetime)
df_ecobike["Hora_Retiro"] = pd.to_datetime(df_ecobike["Hora_Retiro"], format="%H:%M:%S")
df_ecobike["Hora_Arribo"] = pd.to_datetime(df_ecobike["Hora_Arribo"], format="%H:%M:%S")

#Media
media_hora_retiro = df_ecobike["Hora_Retiro"].mean()
media_hora_arribo = df_ecobike["Hora_Arribo"].mean()
print("Media hora de retiro:", media_hora_retiro.strftime('%H:%M:%S'))
print("Media hora de arribo:", media_hora_arribo.strftime('%H:%M:%S'))

#Mediana
mediana_hora_retiro = df_ecobike["Hora_Retiro"].median()
mediana_hora_arribo = df_ecobike["Hora_Arribo"].median()
print("Mediana hora de retiro:", mediana_hora_retiro.strftime('%H:%M:%S'))
print("Mediana hora de arribo:", mediana_hora_arribo.strftime('%H:%M:%S'))

#Moda
moda_hora_retiro = df_ecobike["Hora_Retiro"].mode()[0]
moda_hora_arribo = df_ecobike["Hora_Arribo"].mode()[0]
print("Moda hora de retiro:", moda_hora_retiro.strftime('%H:%M:%S'))
print("Moda hora de arribo:", moda_hora_arribo.strftime('%H:%M:%S'))
```

```python
# Histograma horas de retiro
plt.figure(figsize=(10,5))
plt.hist(df_ecobike["Hora_Retiro"], bins=pd.date_range("1900-01-01 00:00:00", "1900-01-02 00:00:00", freq="1h"), alpha=0.5, edgecolor="black", color="skyblue")

# Líneas de métricas
plt.axvline(media_hora_retiro, color="red", linestyle="dashed", linewidth=2, label="Media")
plt.axvline(mediana_hora_retiro, color="green", linestyle="dashed", linewidth=2, label="Mediana")
plt.axvline(moda_hora_retiro, color="blue", linestyle="dashed", linewidth=2, label="Moda")

plt.text(media_hora_retiro, plt.gca().get_ylim()[1]*0.9, media_hora_retiro.strftime('%H:%M'), color='red',  ha='right')
plt.text(mediana_hora_retiro, plt.gca().get_ylim()[1]*0.85, mediana_hora_retiro.strftime('%H:%M'), color='green')
plt.text(moda_hora_retiro, plt.gca().get_ylim()[1]*0.8, moda_hora_retiro.strftime('%H:%M'), color='blue')

# Eje X en formato horas
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=2))
plt.gcf().autofmt_xdate()

plt.title("Distribución de Horas de Retiro")
plt.xlabel("Hora del día")
plt.ylabel("Cantidad de viajes")
plt.legend()
plt.show()
```

```python
# Histograma horas de arribo
plt.figure(figsize=(10,5))
plt.hist(df_ecobike["Hora_Arribo"], bins=pd.date_range("1900-01-01 00:00:00", "1900-01-02 00:00:00", freq="1h"), alpha=0.5, edgecolor="black", color="skyblue")

# Líneas de métricas
plt.axvline(media_hora_arribo, color="red", linestyle="dashed", linewidth=2, label="Media")
plt.axvline(mediana_hora_arribo, color="green", linestyle="dashed", linewidth=2, label="Mediana")
plt.axvline(moda_hora_arribo, color="blue", linestyle="dashed", linewidth=2, label="Moda")

plt.text(media_hora_arribo, plt.gca().get_ylim()[1]*0.9, media_hora_arribo.strftime('%H:%M'), color='red',  ha='right')
plt.text(mediana_hora_arribo, plt.gca().get_ylim()[1]*0.85, mediana_hora_arribo.strftime('%H:%M'), color='green')
plt.text(moda_hora_arribo, plt.gca().get_ylim()[1]*0.8, moda_hora_arribo.strftime('%H:%M'), color='blue')

# Eje X en formato horas
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=2))
plt.gcf().autofmt_xdate()

plt.title("Distribución de Horas de Arribo")
plt.xlabel("Hora del día")
plt.ylabel("Cantidad de viajes")
plt.legend()
plt.show()
```

## Airbnb Dataset

```python
df_airbnb = pd.read_csv('/content/drive/MyDrive/Colab 6 Semestre/Jesus/listings.csv')
display(df_airbnb.head(20))
```

```python
# Checamos los tipos de datos que tenemos
df_airbnb.info()
```

```python
# Convertimos a valor númerico la columna de "price"
if df_airbnb["price"].dtype == "object":
    df_airbnb["price"] = (df_airbnb["price"].str.replace(r'[$,]', '', regex=True).astype(float))
```

```python
# Definimos las columnas a utilizar
airbnb_data = [
    "price",
    "accommodates", "bedrooms", "beds",
    "review_scores_rating", "review_scores_accuracy", "review_scores_cleanliness",
    "review_scores_checkin", "review_scores_communication",
    "review_scores_location", "review_scores_value", "calculated_host_listings_count",
    "calculated_host_listings_count_entire_homes"
]

# Creamos dataset reducido
df_airbnb_subset = df_airbnb[airbnb_data].copy()
```

### Media

```python
# Media
means = df_airbnb_subset.mean()

print("--Media--")
print(means, "\n")
```

```python
# Graficamos

plt.figure(figsize=(12,6))
ax = means.plot(kind='bar', logy=True, color='skyblue') # Escala logarítmica en el eje y para una mejor visualización de los resultados
plt.title("Media de variables numéricas - Airbnb")
plt.ylabel("Media")
plt.xlabel("Variables")
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(means):
    ax.text(i, v, f"{v:.2f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

### Mediana

```python
# Mediana
medians = df_airbnb_subset.median()

print("--Mediana--")
print(medians, "\n")
```

```python
# Graficamos

plt.figure(figsize=(12,6))
ax = medians.plot(kind='bar', logy=True,  color='skyblue') # Escala logarítmica en el eje y para una mejor visualización de los resultados
plt.title("Mediana de variables numéricas - Airbnb")
plt.ylabel("Mediana")
plt.xlabel("Variables")
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(medians):
    ax.text(i, v, f"{v:.2f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

### Moda

```python
# Moda
modes = df_airbnb_subset.mode().iloc[0] # iloc[0] para asegurarnos de obtener solo 1 valor de moda por cada columna

print("--Moda--")
print(modes)
```

```python
# Graficamos
plt.figure(figsize=(12,6))
ax = modes.plot(kind='bar', logy=True,  color='skyblue') # Escala logarítmica en el eje y para una mejor visualización de los resultados
plt.title("Moda de variables numéricas - Airbnb")
plt.ylabel("Moda")
plt.xlabel("Variables")
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(modes):
    ax.text(i, v, f"{v:.2f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
