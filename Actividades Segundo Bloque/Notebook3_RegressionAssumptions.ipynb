```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
```

```python
# Generate a synthetic dataset
X, y = make_regression(n_samples=200, n_features=1, noise=20, random_state=42)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Fit regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict the values using the test set
y_pred = model.predict(X_test)
```

```python
# Residuals
residuals = y_test - y_pred

# Plot the results
plt.scatter(X_test, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('X_test')
plt.ylabel('Residuals')
```

```python
# Plotting data and regression
plt.figure(figsize=(5,5))
sns.scatterplot(x=X_test.flatten(), y=y_test, label="Current", color='blue')
sns.scatterplot(x=X_test.flatten(), y=y_pred, label="Predicted", color='red')
plt.xlabel('X_test: Independent variable')
plt.ylabel('y_test: Dependent variable')
plt.title('Regression Plot')
plt.legend()
plt.show()

```

```python
# Plot: Homoscedasticity
plt.figure(figsize=(5,5))
sns.scatterplot(x=y_pred, y=np.abs(residuals))
plt.xlabel('Predicted Values')
plt.ylabel('Absolute Residuals')
plt.title('Homoscedasticity plot')
plt.show()
```

```python
# Generate dataset
np.random.seed(42)
X_example = np.random.rand(200, 1) * 100
y_example = 2 * X_example.squeeze() + np.random.normal(0, X_example.squeeze(), 200)

# Split in test and train
X_train, X_test, y_train, y_test = train_test_split(X_example, y_example, test_size=0.3, random_state=42)

# Train a model
model_h = LinearRegression()
model_h.fit(X_train, y_train)
y_pred_h = model_h.predict(X_test)
```

```python
# Plot residual
residuals_h  = y_test - y_pred_h
sns.scatterplot(x=y_pred_h, y=residuals_h)
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=y_pred_h, y=np.abs(residuals_h))
plt.xlabel('Predicted Values')
plt.ylabel('Absolute Residuals')
plt.title('Homoscedasticity Plot')
plt.show()
```

```python
# Plot normality of residuals
plt.figure(figsize=(5,5))
sns.histplot(residuals, kde=True)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```

```python
# Plot Q-Q plot
plt.figure(figsize=(5,5))
sm.qqplot(residuals, line='s')
plt.xlabel('Theoretical Quantities')
plt.ylabel('Sample Quantities')
plt.title('Q-Q Plot')
plt.show()
```

## Wine Dataset

```python
# Wine Dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_data = pd.read_csv(url, delimiter=";")
wine_data.head()
```

```python
# Define variables (X) and (y)
X = wine_data[["alcohol"]]
y = wine_data["quality"]

# Split in test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a model
model_w = LinearRegression()
model_w.fit(X_train, y_train)

# predict the values using the test set
y_pred_w = model_w.predict(X_test)
```

```python
# Residuals
residuals_w = y_test - y_pred_w

# Plot residuals vs alcohol
plt.figure(figsize=(5,5))
plt.scatter(X_test, residuals_w, alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("Alcohol")
plt.ylabel("Residuals")
plt.title("Residuals vs Alcohol")
plt.show()
```

```python
# Plotting data and regression
plt.figure(figsize=(5,5))
sns.scatterplot(x=X_test["alcohol"], y=y_test, label="Actual", color='blue')
sns.scatterplot(x=X_test["alcohol"], y=y_pred_w, label="Predicted", color='red')
plt.xlabel("Alcohol (Independent variable)")
plt.ylabel("Quality (Dependent variable)")
plt.title("Regression Plot: Alcohol vs Wine Quality")
plt.legend()
plt.show()
```

```python
# Plot: Homoscedasticity
plt.figure(figsize=(5,5))
sns.scatterplot(x=y_pred_w, y=np.abs(residuals_w))
plt.xlabel('Predicted Values')
plt.ylabel('Absolute Residuals')
plt.title('Homoscedasticity plot (Wine Dataset)')
plt.show()
```

```python
# Plot normality of residuals
plt.figure(figsize=(5,5))
sns.histplot(residuals_w, kde=True)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals (Wine Dataset)')
plt.show()
```

```python
# Q-Q plot
plt.figure(figsize=(5,5))
sm.qqplot(residuals_w, line='s')
plt.xlabel('Theoretical Quantities')
plt.ylabel('Sample Quantities')
plt.title('Q-Q Plot (Wine Dataset)')
plt.show()
```
